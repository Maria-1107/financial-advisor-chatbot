```python
from pydantic import BaseModel
from typing import Optional
```

```python
class UserProfile(BaseModel):
    name: str
    age: int
    monthly_income: float
    monthly_expense: float
    risk_appetite: str
    investment_goal: str
    existing_investment: Optional[str]=None # optional field

    @property #It lets you access a method like a variable — without calling it with ().
    def monthly_savings(self):
        return self.monthly_income - self.monthly_expense
```

```python
# Create a sample user
user = UserProfile(
    name="Rex",
    age=23,
    monthly_income=50000,
    monthly_expense=30000,
    risk_appetite="medium",
    investment_goal="retirement"
)

print(user.monthly_savings)  # should print 20000.0
```

```python
print(f"Name: {user.name}")
print(f"Age: {user.age}")
print(f"Monthly Income: ₹{user.monthly_income}")
print(f"Monthly Expenses: ₹{user.monthly_expense}")
print(f"Monthly Savings: ₹{user.monthly_savings}")
print(f"Risk Appetite: {user.risk_appetite}")
print(f"Investment Goal: {user.investment_goal}")
```
